```python
text = widgets.Text()

display(text)

def handle_submit(sender):
    print(float(str(text.value)))

text.on_submit(handle_submit,remove=True)
```

```python

val=widgets.FloatSlider(
    value=7.5,
    min=0,
    max=10.0,
    step=0.1,
    description='% Change 2nd Week Retention',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
    slider_color='green'
)
display(val)

def on_value_change(change):
    print(change['new'])

val.observe(on_value_change, names='value')
```
