# Pearson’s Chi-Squared Test of Independence

```python
import scipy
data1=[survey.recall_num, survey.imp_total]
#chi2_contingency(data1)
chi2, p, dof, ex =scipy.stats.chi2_contingency(data1, correction=False)
print 'chi2:', chi2
print 'p-value:', p
print 'degree of freedom:', dof
print 'expected values:', ex
```

# Mann-Whitney-Wilcoxon (MWW) RankSum Test

```python
from scipy import stats  
  
z_stat, p_val = stats.ranksums(survey.recall_num, survey.imp_total)  

print "MWW RankSum P value for Ad Recall & Impression =", p_val  
```

# One-way ANOVA

```python
from scipy import stats  
      
f_val, p_val = stats.f_oneway(survey.recall_num, survey.imp_total)  
  
print "One-way ANOVA P value =", p_val  
```

# Repeat MWW for interest and purchase intent

```python
from scipy import stats  
  
z_stat, p_val = stats.ranksums(survey.interest, survey.imp_total)  
print 'z_stat:', z_stat
print "MWW RankSum P value for Interest & Impression =", p_val  
```

```python
from scipy import stats  
  
z_stat, p_val = stats.ranksums(survey.intent, survey.imp_total)  
print 'z_stat:', z_stat
print "MWW RankSum P value for Purchase Intent & Impression =", p_val  
```

## - Obviously Frequency drives Total Time Viewable, duh?
### --There is clear linear relationship between these 2, shown from spearman correlation matrix heatmap in the beginning

```python
# select only viewable imp and totaol time viewable
# subset_rawData3 = subset_rawData2[['viewableimp','totalviewtime']]
#imp = subset_rawData2['viewableimp']
#vtsec = subset_rawData2['totalviewtime']/1000
# result = pd.concat([df1, df4], axis=1)
#subset_rawData3 = pd.concat([imp,vtsec], axis=1)
# visualize the relationship between viewable imp and total time viewable
# scatter plot
imp = survey['imp_total']
vtsec=survey['vtabove50pct300ms']/1000
survey1 = pd.concat([imp, vtsec], axis=1)
survey1 = survey1[(survey1['imp_total']<=50)&(survey1['vtabove50pct300ms']<=500)]
#survey1.plot(kind='scatter', x='vtabove50pct300ms', y='imp_total', alpha=0.2)
ax=sns.lmplot(x='vtabove50pct300ms', y='imp_total', data=survey1, aspect=1.5,scatter_kws={'alpha':0.2} )
ax.set(xlabel='Time Viewable By Second', ylabel='Number of delivered impressions')
ax.fig.suptitle('Frequency VS Total Time Viewable')
plt.show()
```

### -- Let's take in detail on what happened to total time viewable if the number of impressions are different 

```python
import random
survey[['vtImpMaxAbove50pct300ms', 'vtImpMinAbove50pct300ms',
       'vtImpMeanAbove50pct300ms','vtabove50pct300ms']]=survey[['vtImpMaxAbove50pct300ms', 'vtImpMinAbove50pct300ms',
       'vtImpMeanAbove50pct300ms','vtabove50pct300ms']].applymap(lambda x: random.random() if x==0 else x)
name=list(survey[['vtImpMaxAbove50pct300ms', 'vtImpMinAbove50pct300ms',
       'vtImpMeanAbove50pct300ms','vtabove50pct300ms']].columns.values)
name1=map( lambda m: 'bin-'+m, name) 
survey[name1]=survey[['vtImpMaxAbove50pct300ms', 'vtImpMinAbove50pct300ms',
       'vtImpMeanAbove50pct300ms','vtabove50pct300ms']].apply(lambda x: pd.qcut(x, 10, labels=False), axis=0)
```

```python

vt_imp = pd.DataFrame(pd.pivot_table(survey, values='answersetid', 
                                     index='bin-vtabove50pct300ms',
                    columns=['imp_total'], 
                                     aggfunc=lambda x: len(x.unique())))
vt_imp=vt_imp.ix[:,:5]
vt_imp_pct=vt_imp.apply(lambda c: c / c.sum() * 100, axis=0)
vt_imp_pct=vt_imp_pct
vt_imp_pct.plot(kind='line',linewidth=3)
plt.xlabel('Total Time Viewable Scale 0-9, with 9 being the highest')
plt.ylabel('% of observation in Each Time Scale out of all observations that received same numer of impression')
plt.title('Distribution over Total Time Viewable by number of impressions')
```

# - Frequency - Close - Interest

### -- Frequency VS Close Events
#### Among ads that can be closed, it is likely the more impression being served, more close is triggered

```python
survey_close = survey[(survey.takeover>0)&(survey.feature_swipe==0)]
```

```python
imp1 = survey_close['imp_total']
close =survey_close['close']
survey3 = pd.concat([imp1, close], axis=1)
survey3 = survey3[(survey3['imp_total']<=50)]
ax=sns.lmplot(x='close', y='imp_total', data=survey3, aspect=1.5,scatter_kws={'alpha':0.2} )
ax.set(xlabel='Number of Close', ylabel='Number of delivered impressions')
ax.fig.suptitle('Frequency VS Close Event')
plt.show()
```

### -- Frequency VS % Audience Who do NOT close the Ad

```python
close_imp = pd.DataFrame(pd.pivot_table(survey_close, values='answersetid', 
                                     index='close',
                    columns=['imp_total'], 
                                     aggfunc=lambda x: len(x.unique())))
close_imp=close_imp.fillna(0)
close_imp=close_imp.ix[:,:6]
close_imp_pct=close_imp.apply(lambda c: c / c.sum() * 100, axis=0)
close_imp_pct=close_imp_pct.head(1).reset_index()
close_imp_pct=close_imp_pct[[1,2,3,4]]
close_imp_pct.T.plot(kind='line',linewidth=5)
plt.xlabel('Number of Impressions')
plt.ylabel('% of audience who did NOT close the ad within numer of impression')
plt.title('Distribution audience who did NOT close the ad by number of impressions')
```
