```python
# subset recall
recall = rawData[rawData['questionid']==84].reset_index()
# transform yes and no into numeric representation
# recall['recall'] = np.where(recall['optiontext']=='Yes', 1, 0)
# recall['recall'] = recall.Set.map( lambda x: 'yes' if x == 1 else 0)
recall['recall'] = [1 if x == 'Yes' else 0 for x in recall['optiontext']]
# check if new column recall data type is int
# recall.dtypes
recall= recall [['recall','surveyid','answersetid', 'iscomplete','isrefgroup', 'click', 'firstinteraction', 'deliveredimp', 'viewableimp_ic','totalviewtime' ]]
```

```python
# subset interest
interest = rawData[rawData['questionid']==86].reset_index()
interest['interest'] = pd.to_numeric(interest['optiontext'], errors='coerce')
# interest.dtypes
interest = interest[['interest','surveyid','answersetid', 'iscomplete','isrefgroup', 'click', 'firstinteraction', 'deliveredimp', 'viewableimp_ic','totalviewtime' ]]
```

```python
# subset purchase intent
intent = rawData[rawData['questionid']==83].reset_index()
intent['intent'] = pd.to_numeric(intent['optiontext'], errors='coerce')
intent = intent [['intent','surveyid','answersetid', 'iscomplete','isrefgroup', 'click', 'firstinteraction', 'deliveredimp', 'viewableimp_ic','totalviewtime' ]]
```

```python
# join recall, interest and intent
# result = pd.merge(left, right, how='right', on=['key1', 'key2'])
# df1.merge(df2,on='name').merge(df3,on='name')
df = recall.merge(interest, 
on= ['surveyid','answersetid', 'iscomplete','isrefgroup', 'click', 
     'firstinteraction', 'deliveredimp', 'viewableimp_ic','totalviewtime']).merge(intent, on= ['surveyid','answersetid', 'iscomplete','isrefgroup', 'click', 
     'firstinteraction', 'deliveredimp', 'viewableimp_ic','totalviewtime'])
df.head()
```

```python
# working with null and nan values
# df.totalviewtime.fillna(0, inplace=True)
# df.totalviewtime.notnull()
df.ix[df.totalviewtime == '(null)', 'totalviewtime'] = 0
df['totalviewtime'] = df.totalviewtime.astype(int)
df.dtypes


```

```python
# calculate average view time per viewable impression
df['avgviewtime'] = df['totalviewtime'].div(df['viewableimp_ic'], axis=0)
df.fillna(0, inplace=True)
df.head()
```

```python
# convert view time by second
df['avgviewtime'] = df['avgviewtime']/1000
df['totalviewtime'] = df['totalviewtime']/1000
df.head()
```

```python
from sklearn import metrics
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# define a function that accepts a list of features and returns testing RMSE
def train_test_rmse(feature_cols):
    X = df1[feature_cols]
    y = df1.intent
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    y_pred = linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(y_test, y_pred))


# same true values as above
true = [10, 7, 5, 5]

# new set of predicted values
pred = [10, 7, 5, 13]

# MAE is the same as before
print 'MAE:', metrics.mean_absolute_error(true, pred)

# MSE and RMSE are larger than before
print 'MSE:', metrics.mean_squared_error(true, pred)
print 'RMSE:', np.sqrt(metrics.mean_squared_error(true, pred))
```

```python
print train_test_rmse(['recall','interest'])
```
